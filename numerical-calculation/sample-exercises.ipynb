## numerical solution for diffrential equation via Euler method

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def odeeuler2(f,s0,y0,t):
     s=np.zeros(len(t))
     y=np.zeros(len(t))
    
     s[0]=s0
     y[0]=y0
    
     for n in range(0,len(t)-1):
        s[n+1]=s[n]+f(s[n],y[n],t[n])*(t[n+1]-t[n])
        y[n+1]=y[n]+s[n]*(t[n+1]-t[n])
     return y 
t=np.linspace(0,2*np.pi,500)
y0=1
s0=-0.5
def f(s,y,t):
    return -t-2*y
y=odeeuler2(f,s0,y0,t)
 
 
y_true=np.cos(t*2**(1/2))-t/2
plt.plot(t,y,'r.',t,y_true,'b')
plt.legend(['Euler','true'])
plt.ylabel('y')
plt.xlabel('t')
plt.grid(True)
 
plt.show()
```

## finding roots of a function via Newton method

```python
def f(x):
  return x**5 - x**3 - x
 
x=np.linspace(-2,2,1000)
plt.plot(x,f(x),label='y=f(x)')
plt.legend(loc='best')
plt.grid(True)

def fp(x):
  return 5*x**4 - 3*x**2 -1

#finding roots of f function

x0 = [-5,0.5,5]
prec = 1e-10

for i in range(len(x0)):
  while abs(f(x0[i])) > prec :
     deltax0 = -1 * f(x0[i]) / fp(x0[i])
     x0[i] = x0[i] + deltax0
print('roots of the f function : ',x0)

y0=[0,0,0]
plt.plot(x0,y0,"r*")
  
#finding roots of fp function

def fpp(x):
  return 20*(x**3) - 6*x 
x00 = [-3,2]
for i in range(len(x00)):
  while abs(fp(x00[i])) > prec :
    deltax00 = -1 * fp(x00[i]) / fpp(x00[i])
    x00[i] = x00[i] + deltax00
print('roots of the fp function : ',x00)

y00=[0,0]
plt.plot(x00,y00,'g*')

#numerical calculation of derivative of f function 
y=f(x)
def derivative(f,a,h=0.01):
  return (f(a+h)-f(a-h))/(2*h)
dxdy=derivative(f,x)

plt.plot(x,dxdy,'y',label='Derivative y=f(x)')
plt.legend(loc='best')
plt.xlabel('x')
#parametrical calculation of derivative of f function
from sympy import diff
from sympy.abc import x,y
diff(x**5 - x**3 - x,x)
```

## integrating numerically via Riemann sums method

```python
def f(x):
    return 1/(1+x**2)
a=0 ; b=5; N=10
n=10

x=np.linspace(a,b,N+1)
y=f(x)

X=np.linspace(a,b,n+N+1)
Y=f(X)

plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.plot(X,Y,'b')
```

```python
dx = (b-a)/N
x_left = np.linspace(a,b-dx,N)
x_midpoint = np.linspace(dx/2,b-dx/2,N)
x_right= np.linspace(dx,b,N)

print('partition with',N,'subintervals.')
leftsum=np.sum(f(x_left)*dx)
print('Left Riemann Sum : ', leftsum)
midsum=np.sum(f(x_midpoint)*dx)
print('Midpoint Riemann Sum : ', midsum)
rightsum=np.sum(f(x_right)*dx)
print('Right Riemann Sum : ', rightsum)
```

```python
#error for each method
print('partition with',N,'subintervals.')
leftsum=np.sum(f(x_left)*dx)
print('Error of Left Riemann Sum : ', (leftsum-true) /true)
midsum=np.sum(f(x_midpoint)*dx)
print('Error of Midpoint Riemann Sum : ', (midsum-true)/true)
rightsum=np.sum(f(x_right)*dx)
print('Error of Right Riemann Sum : ', (rightsum-true)/true)
```
